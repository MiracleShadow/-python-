```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime

%matplotlib inline
# %matplotlib notebook
```

```python
hnnu_news             = pd.read_csv('CSVFiles/HNNU_news.csv')
ah_news               = pd.read_csv('CSVFiles/AH_news.csv')
hfut_news             = pd.read_csv('CSVFiles/HFUT_news.csv')
hfut_exchange_news    = pd.read_csv('CSVFiles/HFUT_exchange_news.csv')


hnnu_news['Date']             = pd.to_datetime(hnnu_news['Date'], format='%Y-%m-%d')
ah_news['Date']               = pd.to_datetime(ah_news['Date'], format='%Y-%m-%d')
hfut_news['Date']             = pd.to_datetime(hfut_news['Date'], format='%Y-%m-%d')
hfut_exchange_news['Date']    = pd.to_datetime(hfut_exchange_news['Date'], format='%Y-%m-%d')


print(hnnu_news.dtypes, len(hnnu_news))
print(ah_news.dtypes, len(ah_news))
print(hfut_news.dtypes, len(hfut_news))
print(hfut_exchange_news.dtypes, len(hfut_exchange_news))


hnnu_news['Date_ym'] = hnnu_news['Date'].apply(lambda x: x.year*100+x.month)
ah_news['Date_ym'] = ah_news['Date'].apply(lambda x: x.year*100+x.month)
hfut_news['Date_ym'] = hfut_news['Date'].apply(lambda x: x.year*100+x.month)
hfut_exchange_news['Date_ym'] = hfut_exchange_news['Date'].apply(lambda x: x.year*100+x.month)
# hnnu_news['Date_ym']
```

```python
hnnu_y              = hnnu_news.groupby(['Date_ym']).size()
hnnu_x              = [str(i) for i in list(hnnu_y.index)]

ah_y                = ah_news.groupby('Date_ym').size()
ah_x                = [str(i) for i in list(ah_y.index)]

hfut_y              = hfut_news.groupby('Date_ym').size()
hfut_x              = [str(i) for i in list(hfut_y.index)]

hfut_exchange_y     = hfut_exchange_news.groupby('Date_ym').size()
hfut_exchange_x     = [str(i) for i in list(hfut_exchange_y.index)]
```

```python
plt.figure(figsize=(25,20))

plt.plot(hnnu_x[-26:-1], hnnu_y[-26:-1], c='red', label='hnnu')

plt.plot(ah_x[-26:-1], ah_y[-26:-1], c='green', label='ah')

plt.plot(hfut_x[-26:-1], hfut_y[-26:-1], c='orange', label='hfut')

plt.plot(hfut_exchange_x[-24:-1], hfut_exchange_y[-24:-1]*5, c='black', label='hfut_e')

plt.plot(hfut_x[-26:-1], hfut_exchange_y[-24:-1]+hfut_y[-26:-1], c='yellow', label='hfut_all')

plt.legend(loc='upper left')
plt.xlabel('Month')
plt.ylabel('Number')
plt.title('Number of monthly news')
plt.show()
```
