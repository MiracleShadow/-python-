```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime

%matplotlib inline
```

```python
hnnu_news = pd.read_csv('CSVFiles/HNNU_news.csv')
hnnu_news['Date'] = pd.to_datetime(hnnu_news['Date'], format='%Y-%m-%d')
print(hnnu_news.dtypes, len(hnnu_news))
hnnu_news['Date_ym'] = hnnu_news['Date'].apply(lambda x: x.year*100+x.month)
hnnu_news['Date_ym']
```

```python
y = hnnu_news.groupby(['Date_ym']).size()
x = list(y.index)
x = [str(i) for i in x]
x
```

```python
plt.plot(x, y)
plt.show()
```
