```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime

%matplotlib inline
```

```python
hnnu_news = pd.read_csv('CSVfiles/HNNU_news.csv', header=None)
hnnu_news[1] = pd.to_datetime(hnnu_news[1], format='%Y-%m-%d')
print(hnnu_news.dtypes, len(hnnu_news))
hnnu_news['date'] = hnnu_news[1].apply(lambda x: x.year*100+x.month)
hnnu_news['date']
```

```python
y = hnnu_news.groupby(['date']).size()
x = list(y.index)
x = [str(i) for i in x]
x
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.plot(x, y)
plt.show()
```
